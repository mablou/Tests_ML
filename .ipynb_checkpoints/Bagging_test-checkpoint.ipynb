# Bagging predictors (Breiman, 1996)
Le bagging qui signifie en fait bootstrap aggregating est une méthode pour générer plusieurs versions d'un prédicteur qui peuvent ensuite être combiner en un seul prédicteur moyen avec une variance associée.

1. Prenons une série de données ${(x_1,y_1),...,(x_n,y_n)}$ formée de n observations indépendantes
2. Si on considère qu'on peut estimer les valeurs ${y_1,...,y_n}$ à l'aide d'un estimateur $\phi(x)$ , il existe un ensemble de $k$ prédicteurs issus d'un opération de bootstrapping pour former un opérateur aggregé $\phi_A = \phi(x)^{(k)}$. 

### Un exemple simple de régression linéaire:

```python
np.random.RandomState(123)
x = np.arange(0,100,1)
bruit_blanc = 25 * np.random.normal(size=100)
y = 1.5*x + bruit_blanc
plt.plot(x,y,'o');
```

```python
phi = np.polyfit(x,y,1)
print phi
```

```python
plt.plot(x,y,'o');
plt.plot(x,phi[0]*x+phi[1]);
```

### On peut évaluer la variabilité de notre estimation en faisant du bootstrapping ou plutôt un tirage aléatoire avec remise

```python
indices = np.arange(0,len(x))
n_boot=100
boot_sample_x = np.zeros((len(x),n_boot))
boot_sample_y = np.zeros((len(y),n_boot))
phi_boot = np.zeros((2,100))
for i in range(100):
    id_boot = np.random.choice(indices,len(x))
    boot_sample_x[:,i] = np.array([x[id] for id in id_boot])
    boot_sample_y[:,i] = np.array([y[id] for id in id_boot])
    phi_boot[:,i] = np.polyfit(boot_sample_x[:,i],boot_sample_y[:,i],1)
    plt.plot(x,phi_boot[0,i]*x+phi_boot[1,i]);
    
print np.mean(phi_boot,axis=1)
plt.plot(x,y,'ko');
```
